# Exploratory Data Analysis (EDA) 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')
```

### Customer's Info

```python
print(customers.info(),"\n")
print(customers.head(), "\n")
```

### Product's Info

```python
print(products.info(), "\n")
print(products.head(), "\n")
```

### Transaction's Info 

```python
print(transactions.info(), "\n")
print(transactions.head(), "\n")
```

### Data Cleaning 

```python
print("Checking for missing values:\n")
print("Customers missing values:", customers.isnull().sum(), "\n")
print("Products missing values:", products.isnull().sum(), "\n")
print("Transactions missing values:", transactions.isnull().sum(), "\n")
```

```python
print("Checking for duplicstes:\n")
print("Customers duplicates:", customers.duplicated().sum())
print("Products duplicates:", products.duplicated().sum())
print("Transactions duplicates:", transactions.duplicated().sum())
```

```python
print("Descriptive statistics:\n")

print("Transactions Statistics:\n", transactions.describe(), "\n")
print("Products Statistics:\n", products.describe(), "\n")
```

```python
print("Merging datasets:\n")

merged_data = transactions.merge(customers, on="CustomerID", how="left")
merged_data = merged_data.merge(products, on="ProductID", how="left")
```

```python
merged_data.head()
```

## Visualisation

```python
print("Checking for transactions by region\n")
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_data, x="Region", order=merged_data['Region'].value_counts().index, palette="viridis")
plt.title("Number of Transactions by Region")
plt.ylabel("Number of Transactions")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()
```

we can see that transactions are more in South America

```python
print("Checking for revenues by products")
revenue_by_category = merged_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
revenue_by_category.plot(kind="bar", color="teal")
plt.title("Revenue by Product Category")
plt.ylabel("Revenue (USD)")
plt.xlabel("Product Category")
plt.show()
```

```python
top_customers = merged_data.groupby('CustomerID')['TotalValue'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
top_customers.plot(kind="bar", color="coral")
plt.title("Top 10 Customers by Total Spending")
plt.ylabel("Total Spending (USD)")
plt.xlabel("Customer ID")
plt.show()
```

```python
numeric_cols = ['Quantity', 'TotalValue', 'Price_x']
plt.figure(figsize=(8, 6))
sns.heatmap(merged_data[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Seasonality Analysis (Transaction Volume and Revenue Over Time)
merged_data['TransactionDate'] = pd.to_datetime(merged_data['TransactionDate'])  # Ensure date is in datetime format
merged_data['Month'] = merged_data['TransactionDate'].dt.month
merged_data['Year'] = merged_data['TransactionDate'].dt.year
```

```python
monthly_transactions = merged_data.groupby('Month')['TransactionID'].count()
plt.figure(figsize=(10, 6))
monthly_transactions.plot(kind="line", marker='o', color="purple")
plt.title("Monthly Transaction Volume (Seasonality)")
plt.ylabel("Number of Transactions")
plt.xlabel("Month")
plt.xticks(range(1, 13))
plt.show()
```

```python
monthly_revenue = merged_data.groupby('Month')['TotalValue'].sum()
plt.figure(figsize=(10, 6))
monthly_revenue.plot(kind="line", marker='o', color="green")
plt.title("Monthly Revenue (Seasonality)")
plt.ylabel("Revenue (USD)")
plt.xlabel("Month")
plt.xticks(range(1, 13))
plt.show()
```

```python
# Price-Sensitive vs High-Value Customer Segmentation

merged_data['ProductPrice'] = merged_data['Price_x']
# Segmenting customers into price-sensitive and high-value based on spending habits
customer_spending = merged_data.groupby('CustomerID')['TotalValue'].sum()
```

```python
low_spenders = customer_spending[customer_spending < customer_spending.quantile(0.33)]
high_spenders = customer_spending[customer_spending > customer_spending.quantile(0.67)]
```

```python
# Plotting spending behavior
plt.figure(figsize=(10, 6))
sns.boxplot(x=customer_spending, color="lightblue")
plt.title("Customer Spending Distribution")
plt.xlabel("Total Spending (USD)")
plt.show()
```
